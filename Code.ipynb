```python
#Train and Test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,df.columns!='Outcome']
y=df.iloc[:,df.columns=='Outcome']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)#20%
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
#Algorithm
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
models=RandomForestClassifier()
```

```python
models.fit(xtrain,ytrain.values.ravel()) #train the algorithm
```

```python
predict_output=models.predict(xtest)#test the algorithm
print(predict_output)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(predict_output,ytest)
print('The accuracy score for RandomForest:',accuracy)
```
